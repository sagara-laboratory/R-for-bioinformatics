```r
r_search <- entrez_search(db = "protein", term = "alpha amylase AND Aspergillus oryzae[ORGN]", retmax = NULL)
str(r_search)
```

```r
ProtResult <- entrez_summary(db = "protein", id = r_search$ids)
sapply(ProtResult, "[[", "extra")
```

```r
search_year <- function(year, term){
 query <- paste(term, "AND (", year, "[PDAT])")
 entrez_search(db = "pubmed", term = query, retmax = 0)$count
}
year <- 2005:2022
papers <- sapply(year, search_year, term = "alpha amylase AND Aspergillus oryzae[ORGN]", USE.NAMES = FALSE)
plot(year, papers, type = 'b', main = "alpha amylase AND Aspergillus oryzae[ORGN]:PAPERS", ylim = c(0, type.convert(max(papers))*1.2))
```

```r
SearchResult <- entrez_search(db = "pubmed", retmax = 10, term = "alpha amylase AND Aspergillus oryzae")
GetSummary <- entrez_summary(db="pubmed", id=SearchResult$ids)
sapply(GetSummary, "[[", "title")
```
